```python
from owlready2 import get_ontology
import os

# Load the ontology from local file - allow loading imported ontologies
rdf_path = r"c:\Users\ans83\git\NOWL\nowl-main\test\iof-biopharma\biopharma\AboutBiopharmaceuticalManufacturingDev.rdf"
onto = get_ontology(rdf_path).load()

# Get the specific class
cls_iri = "https://spec.industrialontologies.org/ontology/recipe/Recipe"
cls = onto.world.search_one(iri=cls_iri)

if cls:
    print(f"=== AXIOMS FOR {cls.name} ===\n")
    
    # Equivalent to (Necessary & Sufficient)
    print(f"Equivalent To (Necessary & Sufficient):")
    if cls.equivalent_to:
        for eq in cls.equivalent_to:
            if eq is not None:
                print(f"  {cls.name} EquivalentTo: {eq}")
    else:
        print("  None")
    
    # Is A / SubClassOf (Necessary)
    print(f"\nSubClassOf / Is A (Necessary conditions):")
    if cls.is_a:
        for parent in cls.is_a:
            if parent != cls:  # Avoid self-reference
                print(f"  {cls.name} SubClassOf: {parent}")
    else:
        print("  None")
    
    # Get all superclasses (ancestors)
    superclasses = set(cls.ancestors())
    print(f"\nAll Superclasses (including {cls.name}):")
    for sc in sorted(superclasses, key=lambda x: x.name if hasattr(x, 'name') else str(x)):
        if hasattr(sc, 'name'):
            print(f"  {sc.name}")

    print(f"\nObject properties where {cls.name} or its superclasses are in domain or range:")
    found_props = False
    for prop in onto.world.object_properties():
        found_domain = False
        found_range = False
        domain_class = None
        range_class = None
        
        for dom in prop.domain:
            if dom in superclasses:
                found_domain = True
                domain_class = dom.name if hasattr(dom, 'name') else dom
                break
        
        for rng in prop.range:
            if rng in superclasses:
                found_range = True
                range_class = rng.name if hasattr(rng, 'name') else rng
                break
        
        if found_domain or found_range:
            found_props = True
            parts = []
            if found_domain:
                parts.append(f"domain={domain_class}")
            if found_range:
                parts.append(f"range={range_class}")
            print(f"  {prop.name}: {', '.join(parts)}")
    
    if not found_props:
        print("  None")

    print(f"\nData properties where {cls.name} or its superclasses are in domain:")
    found_data_props = False
    for prop in onto.world.data_properties():
        for dom in prop.domain:
            if dom in superclasses:
                print(f"  {prop.name}: domain={dom.name if hasattr(dom, 'name') else dom}")
                found_data_props = True
                break
    
    if not found_data_props:
        print("  None")
else:
    print(f"Class {cls_iri} not found.")
    print("\nSearching for 'Recipe' class in all loaded ontologies...")
    for cls in onto.world.classes():
        if 'Recipe' in cls.name:
            print(f"  Found: {cls.name} - IRI: {cls.iri}")
```
